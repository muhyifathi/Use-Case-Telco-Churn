```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_telco = pd.read_excel('Telco_customer_churn.xlsx')
```

# Data Cleansing

```python
look_var = ['Monthly Charges', 'Total Charges', 'Total Charges2']
df_telco['Total Charges2'] = df_telco['Total Charges'].astype(str)
df_telco[look_var].sort_values('Total Charges2')[:15]
```

```python
df_telco['Total Charges Clean'] = np.where(df_telco['Total Charges']==' ', df_telco['Monthly Charges'], df_telco['Total Charges'])
df_telco['Total Charges Clean'] = df_telco['Total Charges Clean'].astype(float)
```

# Monetary Measure

```python
services = ['Phone Service',
       'Multiple Lines', 'Internet Service', 'Online Security',
       'Online Backup', 'Device Protection', 'Tech Support', 'Streaming TV',
       'Streaming Movies', 'Contract', 'Paperless Billing', 'Payment Method',
       'Monthly Charges', 'Total Charges']
profile = ['Churn Label', 'Churn Value','Churn Score', 'CLTV', 'Churn Reason']
```

```python
df_telco[services[-2:]+profile]
```

## Churn Rate

```python
churn_rate = round(100*df_telco['Churn Value'].mean(),2)
churn_rate
```

# Monthly Charge Loss

```python
total_monthly_charges = df_telco['Monthly Charges'].sum()
monthly_charges_loss = sum(df_telco['Monthly Charges'] * df_telco['Churn Value'])
monthly_charges_keep = total_monthly_charges - monthly_charges_loss
```

```python
charges = {
    'monthly charges all':total_monthly_charges,
    'monthly charges keep':monthly_charges_keep,
    'monthly charges keep (%)':round(100*monthly_charges_keep/total_monthly_charges,2),
    'monthly charges loss':monthly_charges_loss,
    'monthly charges loss (%)':round(100*monthly_charges_loss/total_monthly_charges,2),
}
charges
```
